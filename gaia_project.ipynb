```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline
#import mesa_reader as mr
import pandas as pd
```

```python
df = pd.read_csv('1526332579027O-result.csv')
br = df.bp_rp
par = df.parallax 
g = df.phot_g_mean_mag
ag = df.a_g_val
ebr = df.e_bp_min_rp_val
temp = df.teff_val

f = pd.read_csv('blue2.csv')
br_f = f.bp_rp
par_f = f.parallax 
g_f = f.phot_g_mean_mag
ag_f = f.a_g_val
ebr_f = f.e_bp_min_rp_val
temp_f = f.teff_val
```

```python
color = br-ebr
mag = g - (5*np.log10(1./par*1000)-5) - ag
lum = 10**(-.4*(mag-5))

color_f = br_f-ebr_f
mag_f = g_f - (5*np.log10(1./par_f*1000)-5) - ag_f
lum_f = 10**(-.4*(mag_f-5))
```
