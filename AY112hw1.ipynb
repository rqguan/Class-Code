```python
#first import our packages
from numpy import *
%matplotlib inline
from matplotlib.pyplot import *
from astropy.io import ascii
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

```python
#define the directory containing your data
# you can define it as an empty string if your files are 
# located in the same folder as your notebook 

# directory = ''  # If files are in the same folder as notebook
directory = '' 

# read in our file -- filename is the string containing the directory
# where the full dataset is, plus the string of the particular filename.

filename = directory +'history-1.data'

#read contents of file into the table "history"
history0 = ascii.read(directory+'history-1.data',header_start=4,data_start=5)
history1 = ascii.read(directory+'history0.data',header_start=4,data_start=5)
history2 = ascii.read(directory+'history3.data',header_start=4,data_start=5)
history3 = ascii.read(directory+'history7.data',header_start=4,data_start=5)
# this command says that we want to start looking for the header *after* 
#   line 4 and for the data *after* line 5 (counting from zero)
```

```python
names = history0.colnames

# we can now print anytime we want, the names of the columns 
print names
```

```python
figure(figsize=(10,6))
plot(history0['star_age'], history0['log_R'], label = "M=0.5 solar mass")
plot(history1['star_age'], history1['log_R'], label = "M=1 solar mass")
plot(history2['star_age'], history2['log_R'], label = "M=8 solar mass")
plot(history3['star_age'], history3['log_R'], label = "M=128 solar mass")
plot([0,0], label = "Made by Runquan")
semilogx() # log x axis Y axis is already log
xlabel('age [yr]')
ylabel('log(Radius) [solar]')
grid()


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
figure(figsize=(10,6))

L = []
M = []
N = []
H = []
for i in history0['star_mass']:
    i = log(i)/log(2)
    L.append(i)

for i in history1['star_mass']:
    i = log(i)/log(2)
    M.append(i)
    
for i in history2['star_mass']:
    i = log(i)/log(2)
    N.append(i)
    
for i in history3['star_mass']:
    i = log(i)/log(2)
    H.append(i)
    

plot(history0['star_age'],L , label = "M=0.5 solar mass")
plot(history1['star_age'],M , label = "M=1 solar mass")
plot(history2['star_age'],N , label = "M=8 solar mass")
plot(history3['star_age'],H , label = "M=128 solar mass")
plot([0,0], label = "Made by Runquan")
semilogx() # log x axis Y axis is already log


xlabel('age [yr]')
ylabel('mass [2^(x) times solar mass]')
grid()


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
